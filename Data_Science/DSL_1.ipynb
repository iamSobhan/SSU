**Python Data Types** <br><br>
<p>In Python, data types specify the kind of value a variable holds and determine what operations can be performed on that data.</p><br>
<p>Python offers several built-in data types for handling different kinds of values, such as numbers, text, collections, and binary data.<br>
Each type enables specialized operations and is essential for effective programming in Python.</p>

**Numeric Types**<br>
<p>Python supports three numeric types: int, float, and complex.</p>

```python
# Integer - whole numbers
a = 10
print(a, type(a))

# Float - decimal numbers
b = 3.14
print(b, type(b))

# Complex - numbers with real and imaginary parts
c = 2 + 3j
print(c, type(c))
```

**String Type** <br>
<p>A string is a sequence of characters enclosed in single, double, or triple quotes.</p>

```python
name = "Python Programming"
print(name, type(name))

# String operations
print(name.upper())      # Convert to uppercase
print(name[0:6])         # Slicing (first 6 characters)
```

**Boolean Type**<br>
<p>Boolean represents truth values — True or False.</p>

```python
x = True
y = False
print(x, type(x))
print(5 > 3)
print(10 == 5)
```

**Sequence Types**<br>
<p>These include list, tuple, and range.</p>

**(A) List** <br>
Ordered, mutable (changeable), allows duplicates.

```python
fruits = ["apple", "banana", "cherry"]
print(fruits, type(fruits))

print(fruits)
fruits.append("orange")
```

**Set Type**<br>
Unordered, mutable, no duplicate elements.


```python
s = {1, 2, 3, 3, 4}
print(s, type(s))   # Output: {1, 2, 3, 4} <class 'set'>

s.add(5)
print(s)            # Output: {1, 2, 3, 4, 5}
```

**a. Local Variables**

Declared inside a function.<br>

Accessible only within that function.

```python
def greet():
    message = "Hello, World!"   # Local variable
    print(message)

greet()
```

**b. Global Variables**<br>
Declared outside any function.

Can be accessed anywhere in the program.

```python
x = 10   # Global variable

def display():
    print("Inside function:", x)

display()
print("Outside function:", x)
```

**c. Global Variable Modified Inside Function** <br>

To modify a global variable inside a function, use the global keyword.

```python
count = 0   # Global variable

def increment():
    global count
    count += 1

increment()
print("Count =", count)
```

**d. Instance Variables** <br>

Belong to objects (instances) of a class.

Defined using self within a class.

```python
class Student:
    def __init__(self, name, age):
        self.name = name    # Instance variable
        self.age = age

s1 = Student("MCA", 20)
print(s1.name, s1.age)
```

**e. Class Variables**

Shared among all instances of a class.

Defined inside a class but outside methods.

```python
class Student:
    college = "Seacom Skills University"   # Class variable

    def __init__(self, name):
        self.name = name         # Instance variable

s1 = Student("BCA")
s2 = Student("MCA")

print(s1.name, s1.college)
print(s2.name, s2.college)
```

**Python Operators**<br><br>
<p>Python Operators are special symbols or keywords used to perform operations on values or variables.
They tell Python to perform mathematical, logical, comparison, or assignment actions.</p><br>
<p>Arithmetic Operators<br>
Assignment Operators<br>
Comparison / Relational Operators<br>
Logical Operators<br>
Bitwise Operators<br>
Membership Operators<br>
Identity Operators</p>

**Arithmetic Operators**<br>
Arithmetic operators are used for all basic mathematical operations.

```python
a = 15
b = 4

print("Addition:", a + b)
print("Subtraction:", a - b)
print("Multiplication:", a * b)
print("Division:", a / b)
print("Floor Division:", a // b)
print("Modulus:", a % b)
print("Exponentiation:", a ** b)
```

**Assignment Operators**<br>
Assignment operators update variable values using other operators in a single step.

```python
a = 10
a += 5  # a becomes 15
a *= 2
print(a)
```

**Comparison Operators**<br>
Comparison operators evaluate the relationship between values, returning True or False.

```python
print(5 == 5)
print(10 > 20)
print(8 != 3)
```

**Logical Operators**<br>
Logical operators combine boolean values for control flow and conditions.

```python
a = 10
b = 20

print(a > 5 and b > 15)
print(a > 15 or b > 15)
print(not(a > 5))
```

**Bitwise Operators**<br>
Bitwise operators manipulate bits directly; mainly useful for low-level programming.

```python
x = 5
y = 3
print(x & y)
print(x | y)
print(x ^ y)
print(~x)
print(x << 2)
print(x >> 1)
```

**Identity Operators**<br>
Identity operators test if two references point to the same object in memory.

```python
x = [1, 2, 3]
y = x
z = [1, 2, 3]

print(x is y)
print(x is z)
print(x == z)
```

**Membership Operators**<br>
Membership operators check if a value exists in a collection like list, tuple, set, or string.

```python
fruits = ["apple", "banana", "cherry"]

print("apple" in fruits)
print("orange" not in fruits)
```
