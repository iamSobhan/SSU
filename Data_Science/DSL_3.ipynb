**Series**
<p>A one-dimensional labeled array, similar to a column in Excel.</p>

```python
#Creating a Series(1D Data)
import pandas as pd

s = pd.Series([10, 20, 30, 40])
print(s)
```

**Series from a List of Strings**

```python
names = ["Virat", "Sachin", "Rohit"]
series_names = pd.Series(names, name = "Player Names")
print(series_names)
```

**Series from a List of Integers**

```python
runs = [85, 92, 78]
series_runs = pd.Series(runs, name = "Runs")
print(series_runs)
```

**Series from a Custom List**

```python
players = [("Virat", 85), ("Sachin", 92), ("Rohit", 78)]
series_players = pd.Series(players, name="Player (Name, Runs)")
print(series_players)
```

**Setting a Custom Index and Name**

```python
series_custom = pd.Series(runs, index=names, name="Player Scores")
print(series_custom)
```

**Series from Dictonary**

```python
data = {"a": 10, "b": 20, "c": 30}
s = pd.Series(data)
print(s)
```

**Series from dict with custom index order**

```python
data = {"a": 10, "b": 20, "c": 30}
s = pd.Series(data, index=["b", "c", "a"])
print(s)
```

```python
data = {"a": 10, "b": 20, "c": 30}
s = pd.Series(data, index=["b", "c", "d", "a"])
print(s)
```

**Setting a name for the Series**

```python
subjects = {"Math": 85, "Science": 92, "English": 78}
marks = pd.Series(subjects, name="Student Marks")
print(marks)
print("Series name:", marks.name)
```

```python
#index
#It gives the index labels of the Series
print(s.index)
```

```python
#index values
print(list(s.index))
```

```python
#values
#It returns the Series values as a NumPy array
print(s.values)
```

**Pandas DataFrame**<br>
<p>A two-dimensional, size-mutable, and heterogeneous tabular data structure with labeled axes (rows and columns).</p>

```python
#Creating a DataFrame(2D Table)

data = {
    "Name": ["Virat", "Sachin", "Rohit"],
    "Age": [37, 52, 38],
    "Runs": [85, 90, 95]
}

df = pd.DataFrame(data)
print(df)
```

```python
#printing last 5 rows
df.tail()
```

```python
#printing info about columns, data types, nulls
df.info()
```

```python
#printing summary for numerical columns
df.describe()
```

```python
#printing rows & columns count
df.shape
```

```python
#printing list of column names
df.columns
```

```python
#printing data types of each column
df.dtypes
```

**Selecting Data (Indexing)**

```python
#selecting a single column
df["Name"]
```

```python
#selecting multiple columns
df[["Name", "Runs"]]
```

```python
#selecting row by label
df.loc[1]
```

**Filtering Rows**

```python
#using conditions
df[df["Runs"] > 90]
```

**Adding & Removing Columns**

```python
#adding new column
df["Strike Rate"] = ["106.25", "122", "120"]
```

```python
print(df)
```

```python
#dropping a column
df.drop("Strike Rate", axis=1, inplace=True)
```

```python
print(df)
```

**Adding & Removing Rows**

```python
#adding a new row
new_row = {"Name": "David", "Age": 22, "Runs": 88}

#Converting the new row to a DataFrame before concatenating
new_row_df = pd.DataFrame([new_row])
df = pd.concat([df, new_row_df], ignore_index=True)
```

```python
print(df)
```

```python
#dropping row by index
df.drop(1, axis=0, inplace=True)
```

```python
print(df)
```

**Handling Missing Data**

```python
#filling missing values
df.fillna(0)
```

```python
#dropping rows with missing values
df.dropna()
```

**Sorting Data**

**Grouping & Aggregation**

```python
#group by a column
df.groupby("Age")["Runs"].mean()
```

**Merging, Joining, Concatenating DataFrames**

```python
data = {
    "Name": ["Hardik", "Subhman", "Sreyas"],
    "Age": [30, 28, 29],
    "Runs": [85, 90, 95]
}

df1 = pd.DataFrame(data)
print(df1)
```

```python
data = {
    "Name": ["Rahul", "Bumrah", "Jadeja"],
    "Age": [34, 30, 35],
    "Runs": [63, 12, 88]
}

df2 = pd.DataFrame(data)
print(df2)
```

```python
#concatenate
pd.concat([df1, df2])
```

```python
#merge (SQL JOIN)
pd.merge(df1, df2, on="Age")
```

```python
#join
pd.DataFrame(df1).join(pd.DataFrame(df2), lsuffix="_df1", rsuffix="_df2")
```
