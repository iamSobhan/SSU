**Data Transformation** <br>
<p> Data Transformation is the process of converting raw data into a suitable format or structure so that Machine Learning, Deep Learning, and Data Mining algorithms can learn effectively and produce accurate results. <br>
It mainly includes: <br>

*   Feature Scaling (Normalization / Standardization)
*   Encoding Categorical Data

It's needed because,

*   ML models work with numbers, not text.
*   Features must be on comparable scales.
*   Improves model accuracy and convergence.</p>





**Feature Scaling**<br>
**Min-Max Normalization**

```python
#Min–Max Normalization on a Single Column

import numpy as np
import pandas as pd
```

```python
data = pd.DataFrame({
    "Age": [18, 22, 25, 30, 35, 40, 45, 50]
})

print(data)
```

```python
data["Age_minmax"] = (
    (data["Age"] - data["Age"].min()) /
    (data["Age"].max() - data["Age"].min())
)

print(data)
```

```python
#Min–Max Normalization on Multiple Columns

data = pd.DataFrame({
    "Age": [18, 22, 25, 30, 35],
    "Fare": [100, 250, 400, 800, 1200]
})

data_minmax = (data - data.min()) / (data.max() - data.min())

print(data_minmax)
```

```python
#Using sklearn
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()

data_scaled = scaler.fit_transform(data)

data_scaled = pd.DataFrame(
    data_scaled,
    columns=data.columns
)

print(data_scaled)
```

```python
#With Custom Range
scaler = MinMaxScaler(feature_range=(0, 10))
data_scaled = scaler.fit_transform(data)

print(data_scaled)
```

**Z-score Normalization**

```python
#on a Single Column
data2 = pd.DataFrame({
    "Age": [18, 22, 25, 30, 35, 40, 45, 50]
})

print(data2)
```

```python
mean_age = data2["Age"].mean()
std_age = data2["Age"].std()

data2["Age_zscore"] = (data2["Age"] - mean_age) / std_age

print(data2)
```

```python
#on Multiple Columns
data3 = pd.DataFrame({
    "Age": [18, 22, 25, 30, 35],
    "Fare": [100, 250, 400, 800, 1200]
})

data_zscore = (data3 - data3.mean()) / data3.std()

print(data_zscore)
```

```python
#Using sklearn
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()

data_scaled = scaler.fit_transform(data2)

data_scaled = pd.DataFrame(
    data_scaled,
    columns=data2.columns
)

print(data_scaled)
```

**One-Hot Encoding**

```python
#on Single Categorical Column

data4 = pd.DataFrame({
    "Gender": ["Male", "Female", "Female", "Male", "Other"]
})

print(data4)
```

```python
encoded_data = pd.get_dummies(data4, columns=["Gender"])
print(encoded_data)
```

```python
#with Multiple Categorical Columns

data5 = pd.DataFrame({
    "City": ["Delhi", "Mumbai", "Delhi", "Kolkata"],
    "Department": ["HR", "IT", "Finance", "IT"]
})

encoded_data = pd.get_dummies(data5)
print(encoded_data)
```

```python
#using sklearn
from sklearn.preprocessing import OneHotEncoder
```

```python
data6 = pd.DataFrame({
    "Color": ["Red", "Blue", "Green", "Red"]
})

encoder = OneHotEncoder(sparse_output=False)

encoded = encoder.fit_transform(data6[["Color"]])

encoded_df = pd.DataFrame(
    encoded,
    columns=encoder.get_feature_names_out(["Color"])
)

print(encoded_df)
```

**Applying Normalization & One-Hot Encoding on Titanic Dataset**

```python
titanic = pd.read_csv("/content/titanic.csv")
```

```python
titanic.head()
```

```python
titanic.columns
```

```python
titanic.info()
```

```python
titanic.isnull().sum()
```

```python
titanic["Age"] = titanic["Age"].fillna(titanic["Age"].median())
titanic["Cabin"] = titanic["Cabin"].fillna(titanic["Cabin"].mode()[0])
titanic["Embarked"] = titanic["Embarked"].fillna(titanic["Embarked"].mode()[0])
```

```python
titanic.isnull().sum()
```

```python
titanic.head()
```

```python
#Feature Scaling using Pandas
#Min-Max Normalization

titanic["Age_minmax"] = (titanic["Age"] - titanic["Age"].min()) / \
                     (titanic["Age"].max() - titanic["Age"].min())

titanic["Fare_minmax"] = (titanic["Fare"] - titanic["Fare"].min()) / \
                      (titanic["Fare"].max() - titanic["Fare"].min())

titanic[["Age", "Age_minmax", "Fare", "Fare_minmax"]].head()
```

```python
#Feature Scaling using Scikit-learn
#Min-Max Scaling

from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

titanic[["Age_scaled", "Fare_scaled"]] = scaler.fit_transform(
    titanic[["Age", "Fare"]]
)

titanic.head()
```

```python
#Z-Score Normalization
from sklearn.preprocessing import StandardScaler

std_scaler = StandardScaler()

titanic[["Age_std", "Fare_std"]] = std_scaler.fit_transform(
    titanic[["Age", "Fare"]]
)

titanic.head()
```

```python
#Label Encoding
titanic["Sex_label"] = titanic["Sex"].map({"male": 1, "female": 0})
titanic[["Sex", "Sex_label"]].head()
```

```python
titanic["Embarked_label"] = titanic["Embarked"].map({"S": 0, "C": 1, "Q": 2})
titanic[["Embarked", "Embarked_label"]].head()
```

```python
#Encoding using Scikit-learn
#One-Hot Encoding

from sklearn.preprocessing import OneHotEncoder

encoder = OneHotEncoder(sparse_output=False)

encoded = encoder.fit_transform(titanic[["Sex", "Embarked"]])

encoded_df = pd.DataFrame(
    encoded,
    columns=encoder.get_feature_names_out(["Sex", "Embarked"])
)

encoded_df.head()
```
